```python
from transformers import BertTokenizer, BertModel

model_name = "bert-base-uncased"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv("../../../data/health_fact/clean_dataset/train.csv") # dataset only removed nan claim, label (true and false o only) and main text
```

```python
X = df["claim"]
Y = df["label"]

X_split = np.array_split(X, 3)
Y_split = np.array_split(Y, 3)

from collections import Counter
var = Counter(Y)
```

```python

# embeddings = np.array(embeddings).reshape(-1, embeddings.shape[-1])

# np.save("../../data/embeddings-3", embeddings)

embeddings = np.load("../../../data/health_fact/embeddings/embeddings_clean_train.npy")

print(embeddings.shape)
```

```python
from sklearn.cluster import KMeans

num_clusters = 2  # Adjust this as needed
kmeans = KMeans(n_clusters=num_clusters)
cluster_assignments = kmeans.fit_predict(embeddings)
```

reduce dimensionality

```python
from sklearn.manifold import TSNE

n_components = 2

# Create a t-SNE model
tsne = TSNE(n_components=n_components, random_state=42)

# Fit the model and transform the data
reduced_data = tsne.fit_transform(embeddings)
```

```python
import plotly.express as px

df_show = pd.DataFrame()
df_show["X"] = reduced_data[:, 0]
df_show["Y"] = reduced_data[:, 1]
# df_show["Z"] = reduced_data[:, 2]
df_show["cluster"] = cluster_assignments
df_show["label"] = Y

colors = {
    0: 'red',
    1: 'blue',
    2: 'green'
}

df_show['color'] = df_show['cluster'].map(colors)

fig = px.scatter(df_show, x='X', y='Y', color='label', title='Scatter Plot with Different Cluster Colors')
fig.show()
```

```python
n_components = 3

# Create a t-SNE model
tsne = TSNE(n_components=n_components, random_state=42)

# Fit the model and transform the data
reduced_data = tsne.fit_transform(embeddings)

```

```python
df_show3d = pd.DataFrame()
df_show3d["X"] = reduced_data[:, 0]
df_show3d["Y"] = reduced_data[:, 1]
df_show3d["Z"] = reduced_data[:, 2]
df_show3d["cluster"] = cluster_assignments
df_show3d["label"] = Y

colors = {
    0: 'red',
    1: 'blue',
    2: 'green'
}

df_show3d['color'] = df_show3d['cluster'].map(colors)

fig = px.scatter_3d(df_show3d, x='X', y='Y', z="Z", color='label', title='Scatter Plot with Different Cluster Colors')
fig.show()
```
