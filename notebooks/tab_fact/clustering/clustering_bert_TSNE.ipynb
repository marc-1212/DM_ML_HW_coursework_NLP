```python
import numpy as np

statement = np.load("./statement_npy.npy")

table_text = np.load("./table_text_npy.npy")
```

```python
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE

tsne = TSNE(n_components=3, random_state=42)
data_3d = tsne.fit_transform(statement)
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(data_3d)
```

```python
%pip install matplotlib
```

```python
import matplotlib.pyplot as plt

labels_3d = kmeans.labels_
centroids_3d = kmeans.cluster_centers_
# Crear la figura
fig = plt.figure(figsize=(18, 6))

# Subgráfico 3D con t-SNE y K-Means
ax1 = fig.add_subplot(131, projection='3d')
ax1.scatter(data_3d[:, 0], data_3d[:, 1], data_3d[:, 2], c=labels_3d, cmap='viridis', edgecolor='k', s=50, label='Clusters')
ax1.scatter(centroids_3d[:, 0], centroids_3d[:, 1], centroids_3d[:, 2], c='red', marker='x', s=200, label='Centroides')
ax1.set_xlabel('Dimensión 1 (t-SNE)')
ax1.set_ylabel('Dimensión 2 (t-SNE)')
ax1.set_zlabel('Dimensión 3 (t-SNE)')
ax1.set_title('K-Means Clustering en Datos 3D con t-SNE')
ax1.legend()

# Ajustar el diseño para evitar superposiciones
plt.tight_layout()

# Mostrar los gráficos
plt.show()
```
