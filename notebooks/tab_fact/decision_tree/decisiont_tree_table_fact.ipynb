```python
import numpy as np
import pandas as pd
from datasets import load_dataset
file_npy_statement = '/content/drive/MyDrive/DATA_MINING_COURSEWORK/TAB_FACT_EMBEDDINGS/statement_npy.npy'
file_npy_table_text = '/content/drive/MyDrive/DATA_MINING_COURSEWORK/TAB_FACT_EMBEDDINGS/tablte_text_npy.npy'

data = load_dataset("tab_fact", "tab_fact")
```

```python
X1, X2 = np.load(file_npy_statement), np.load(file_npy_table_text)
```

```python


train_input = pd.DataFrame()

# train_input["X1"] = X1
# train_input["X2"] = X2
train_input["label"] = Y = load_dataset("tab_fact", "tab_fact")["train"].to_pandas()["label"]

train_input.head()
```

```python
from sklearn import tree

input = np.concatenate((X1, X2), axis=1)


clf  = tree.DecisionTreeClassifier()

clf.fit(input, load_dataset("tab_fact", "tab_fact")["train"]["label"])
```

```python
import os
folder_path_statement = '/content/drive/MyDrive/DATA_MINING_COURSEWORK/TAB_FACT_EMBEDDINGS/test/statement'
folder_path_table_text = '/content/drive/MyDrive/DATA_MINING_COURSEWORK/TAB_FACT_EMBEDDINGS/test/table_text'

files_table_text = [f1 for f1 in os.listdir(folder_path_table_text) if os.path.isfile(os.path.join(folder_path_table_text, f1))]
files_statement = [f2 for f2 in os.listdir(folder_path_statement) if os.path.isfile(os.path.join(folder_path_statement, f2))]
```

```python
import numpy as np

statement_vector = []
table_text_vector = []

for i in range(len(files_statement)):
  statement_vector.append(np.load(f"{folder_path_statement}/{files_statement[i]}"))
  table_text_vector.append(np.load(f"{folder_path_table_text}/{files_table_text[i]}"))


embeddings_statement = np.concatenate(statement_vector)
embeddings_table_textt = np.concatenate(table_text_vector)
```

```python
file_npy_statement = '/content/drive/MyDrive/DATA_MINING_COURSEWORK/TAB_FACT_EMBEDDINGS/test_statement_npy'
file_npy_table_text = '/content/drive/MyDrive/DATA_MINING_COURSEWORK/TAB_FACT_EMBEDDINGS/test_tablte_text_npy'
```

```python
np.save(file_npy_statement, embeddings_statement)
np.save(file_npy_table_text, embeddings_table_textt)
```

```python
input_test = np.concatenate((embeddings_statement, embeddings_table_textt), axis=1)
```

```python
output_test = load_dataset("tab_fact", "tab_fact")["test"]["label"]
```

```python
output_pred = clf.predict(input_test)
```

```python
total_samples = len(output_test)
correct = 0

for i in range(total_samples):
  if output_pred[i] == output_test[i]:
    correct += 1

print (correct / total_samples)
```
